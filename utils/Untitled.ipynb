```kotlin
%use lets-plot

import kotlin.math.*
import org.jetbrains.letsPlot.intern.Plot
```

```kotlin
fun meshgridPoints(x:List<Double>, y:List<Double>):List<Pair<Double, Double>> {
  val xSer = y.flatMap { x }
  var yInd = -1
  val ySer = y.flatMap { 
      yInd++
      List(x.size) {y[yInd]} 
  }
  return xSer.zip(ySer)
}

val gridPoints = meshgridPoints(
    generateSequence(-4.0, {it + 15.0/50} ).takeWhile { it <= 0.0 }.toList(),
    generateSequence(-3.0, {it + 15.0/40} ).takeWhile { it <= 0.0 }.toList())
```

```kotlin
val ff = fun(x1: Double, x2: Double) : Double { return x1 * x1 + 2 * x2 * x2 }

val X = gridPoints.map { it.first }
val Y = gridPoints.map { it.second }
val Z = gridPoints.map { ff(it.first, it.second) }
```

```kotlin
val p = letsPlot {x=X; y=Y}
p + geomContour(color="red", binWidth = 5.0) {z=Z}
```

```kotlin
DISPLAY("afo")
```

```kotlin
fun afo() : Plot {
   val p = letsPlot {x=X; y=Y}
   return p+ geomContour(color="red", binWidth = 5.0) {z=Z}
}
```
